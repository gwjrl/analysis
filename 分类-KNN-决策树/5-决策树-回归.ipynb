```python
import numpy as np

from sklearn.tree import DecisionTreeRegressor

import matplotlib.pyplot as plt
```

```python
X = np.linspace(0,2*np.pi,10).reshape(-1,1)

y = np.concatenate([np.sin(X),np.cos(X)],axis = 1)
display(X.shape,y.shape)
```

```python
plt.scatter(y[:,0],y[:,1])
plt.axis('equal')
```

```python
X_test = np.linspace(0,2*np.pi,256).reshape(-1,1)

X_test.shape
```

```python
dtr = DecisionTreeRegressor()

# 算法，训练，建立关系，X（40,1） ----- y（40，2）
dtr.fit(X,y)
```

```python
# 预测 X_test (256,1) --------> y_ (256,2)
y_ = dtr.predict(X_test)
# （256,2）
y_.shape
```

```python
plt.scatter(y_[:,0],y_[:,1])
plt.axis('equal')
```

```python
from sklearn import tree
```

```python
plt.figure(figsize=(24,12))
_ = tree.plot_tree(dtr)
plt.savefig('./dtr.png',dpi = 100)
```
