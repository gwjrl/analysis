* 计算数据总的散度矩阵S_t
* 计算数据类内散度矩阵S_w
* 计算类间散度矩阵S_b
* 特征值和特征向量计算
* 筛选特征向量，进行矩阵运算返回输出结果

```python
import warnings
warnings.filterwarnings('ignore')
import numpy as np

from sklearn.discriminant_analysis import LinearDiscriminantAnalysis

from sklearn import datasets

# 线性代数，功能更加强大
from scipy import linalg
```

```python
X,y = datasets.load_iris(True)
```

```python
X.shape[1]
```

```python
np.unique(y)
```

```python
min(3 - 1, X.shape[1])
```

```python
# eigen 特征值
lda = LinearDiscriminantAnalysis(n_components=4,solver='eigen')

lda.fit(X,y)

X_lda = lda.transform(X)
X_lda[:5]
```

```python
# 第一步，总的散度矩阵total,协方差
S_t = np.cov(X.T,bias = 1)
S_t
```

```python
# 第二步，类内散度矩阵 within
S_w = np.zeros_like(S_t)
for i in range(3):
    cond = y == i
    S_w += np.cov(X[cond],rowvar=False,bias = True)
S_w/=3
S_w
```

```python
# 第三步，类间的散度矩阵
S_b = S_t - S_w
S_b
```

```python
# 第四步，特征特征向量
# scipy 线性代数模块计算
evals, evecs = linalg.eigh(S_b, S_w)
display(evals,evecs)
```

```python
X_lda[:5]
```

```python
# 第五步，获取特征所对应的特征向量
evecs = evecs[:,np.argsort(evals)[::-1]]
X.dot(evecs)[:,:2]
```
