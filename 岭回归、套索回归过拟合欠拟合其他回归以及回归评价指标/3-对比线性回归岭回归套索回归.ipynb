```python
import numpy as np

from sklearn.linear_model import LinearRegression,Ridge,Lasso
```

```python
# Lasso回归用于解决稀松矩阵类型的数据，解（过拟合）
# Ridge回归，过拟合，系数衰减
```

```python
# X是数据
# 50 样本 ----->方程
# 200 特征 ---->未知数 w
# 200元一次方程，50样本，50个方程
# 无解
# 无解 无穷多个解
X = np.random.randn(50,200)
X.shape
X
```

```python
x + y + z = 10
x + 2y + 3z = 20
```

```python
# w 稀松矩阵，因为190个是0，权重是0，不重要，10是非零，有一定的权重，重要
w = np.random.randn(200)
index = np.arange(200)
np.random.shuffle(index)
w[index[:190]] = 0
y = X.dot(w)
# 50样本 ----->目标值y.shape (50,)
y.shape
```

```python
import matplotlib.pyplot as plt
```

```python
linear_model = LinearRegression(fit_intercept=False)

ridge_model = Ridge(alpha= 100,fit_intercept=False)

lasso_model = Lasso(alpha=0.1,fit_intercept=False)

linear_model.fit(X,y)

ridge_model.fit(X,y)

lasso_model.fit(X,y)

plt.figure(figsize=(10,6))
ax = plt.subplot(2,2,1)
ax.plot(w)

ax = plt.subplot(2,2,2)
ax.plot(linear_model.coef_)
ax.set_title('linear')

ax = plt.subplot(223)
ax.plot(ridge_model.coef_)
ax.set_title('ridge')

ax = plt.subplot(224)
ax.plot(lasso_model.coef_)
ax.set_title('lasso')
```
