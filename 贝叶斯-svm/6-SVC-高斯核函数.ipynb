```python
import numpy as np

import matplotlib.pyplot as plt

from sklearn.svm import SVC
```

```python
X = np.random.randn(300,2)
plt.scatter(X[:,0],X[:,1])
```

```python
y = X[:,0]*X[:,1] >=0
y
```

```python
plt.scatter(X[:,0],X[:,1],c = y)
```

```python
x1 = np.linspace(-3,3,200)

y1 = np.linspace(-3,3,180)
# print(x1,y1,sep ='\n',end = '\n-----------------------------\n')
X1,Y1 = np.meshgrid(x1,y1)

# print(X1,Y1,sep = '\n')

plt.scatter(X1,Y1)
```

```python
X_test = np.concatenate([X1.reshape(-1,1),Y1.reshape(-1,1)],axis = 1)
X_test.shape
```

```python
svc = SVC(kernel = 'rbf')

svc.fit(X,y)

y_ = svc.predict(X_test)

plt.figure(figsize=(5,5))

plt.scatter(X_test[:,0],X_test[:,1],c = y_)
```

### 画图查看原理

```python
# 二维 --------> 三维
# 分离平面，上面一类点，下面也是一类点
# 这些点有的距离近，有的距离远
d = svc.decision_function(X_test)
d
```

```python
print(d.max(),d.min())
```

```python
X1.shape
```

```python
# 等高线
plt.figure(figsize=(5,5))
plt.contour(X1,Y1,d.reshape(180,200))
```

```python
# 等高面
plt.figure(figsize=(5,5))
plt.contourf(X1,Y1,d.reshape(180,200))
```
