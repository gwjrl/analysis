```python
import numpy as np

import matplotlib.pyplot as plt
```

```python
# 条形图
x = np.arange(1,13)
h = np.random.randint(10,30,size = 12)
# plt.bar(x,h)
# plt.barh(x,h,color = ['red','green','blue'])
plt.bar(x,h,color = np.random.rand(12,3))
```

```python
# 直方图
x = np.random.randint(0,10,size = 20)
plt.hist(x,bins = 5,rwidth = 0.5)
```

```python
plt.hist(x,bins = 5,rwidth = 0.5,range = [0,10],density=True)
```

```python
# 散点图

x = np.random.randint(0,10,size = 30)

y = np.random.randint(0,10,size = 30)

plt.scatter(x,y,marker = '*',s = 100,c = np.random.rand(30,3))
```

```python
#面积图 Draw a stacked area plot.
# Stacked 堆叠
import matplotlib.pyplot as plt
days = [1,2,3,4,5]  
sleeping =[7,8,6,11,7]
eating = [2,3,4,3,2]
working =[7,8,7,2,2]
playing = [8,5,7,8,13] 
plt.stackplot(days, sleeping,eating,working,playing, colors=['m','c','r','k'])  
plt.xlabel('x')
plt.ylabel('y')
plt.title('Stack Plot')
plt.legend(['Sleeping','Eating','Working','Playing'])
```

```python
plt.rcParams['font.size'] = 20
```

```python
# 饼图
plt.figure(figsize=(9,9))
percent = [0.3,0.225,0.125,0.1,0.2,0.05]
_ = plt.pie(percent,autopct='%0.2f%%',explode=[0,0,0.1,0,0,0.2],
            labels=['Dog','Cat','pig','rat','rabbit','snake'],shadow=True)
```
