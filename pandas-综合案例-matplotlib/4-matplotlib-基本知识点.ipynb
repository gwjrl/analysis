```python
import numpy as np

import matplotlib.pyplot as plt
```

```python
def f(t):    
    return np.exp(-t) * np.cos(2*np.pi*t)
```

```python
# 网格线
x = np.linspace(0,2*np.pi,100)

y = f(x)

plt.plot(x,y)
plt.grid(True,color = 'red',alpha = 0.3,linestyle = '--')

plt.savefig('c.pdf',dpi = 100)
```

$$\frac{\pi}{2}$$

```python
plt.figure(figsize=(12,9))
plt.plot(x,np.sin(x))

plt.plot(x,np.cos(x))

plt.legend(['sin(x)','cos(x)'],loc = 1)

plt.yticks([-1,0,1],labels=['min',0,'max'],size = 20)

_ = plt.xticks(np.arange(0,2*np.pi+0.01,np.pi/2),[0,'$\pi/2$','$\pi$','3$\pi$/2','$2\pi$'])
```

```python
plt.figure(figsize=(12,6))
ax = plt.subplot(121,facecolor = 'green')

ax.plot(x,np.sin(x),color = 'red')

ax = plt.subplot(122,facecolor = 'red')
ax.plot(x,np.cos(x),color = 'white')

plt.savefig('./e.png',dpi = 100,facecolor = 'purple')
```
