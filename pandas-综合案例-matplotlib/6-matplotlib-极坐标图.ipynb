```python
import numpy as np

import matplotlib.pyplot as plt
```

```python
# 极坐标图就是条形图
x = np.arange(0,2*np.pi,np.pi/4)
y = np.random.randint(5,50,size = 8)
y
```

```python
plt.bar(x,y)
```

```python
ax = plt.subplot(111,polar = True)

ax.bar(x,y,color = np.random.rand(8,3),width = 0.5)
```

```python
ax = plt.subplot(111,polar = True)
x = np.arange(0, 6*np.pi, 0.01)
y = 2 * np.pi * x
ax.plot(x,y)
```

```python
x = np.linspace(0,4*np.pi,200)

y = np.exp(x)

plt.plot(x,y)

# plt.yscale('log')
```

```python
x = np.linspace(0,12*np.pi,200)

y = np.exp(x)

ax = plt.subplot(111,projection='polar')

ax.plot(x,y)

# ax.set_yscale('log')
```
