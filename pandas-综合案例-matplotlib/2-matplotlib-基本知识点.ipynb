```python
import numpy as np

import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,2*np.pi,30)
y = np.sin(x)

# plt.plot(x,y,c = 'purple')
# plt.plot(x,y,c = '#FFae83')
# plt.plot(x,y ,c = np.random.rand(3))
# plt.plot(x,y,linestyle = '--')
plt.plot(x,y,marker = '*',markersize = 20,markerfacecolor = 'red',markeredgecolor = 'r')
```

```python
angle = np.linspace(0,2*np.pi,100)

x = np.sin(angle)
y = np.cos(angle)

plt.plot(x,y)

# plt.axis([-1.5,1.5,-1,1])#刻度的设置
_ = plt.axis('equal')

plt.xlabel('X',size = 30)
plt.ylabel('Y',size = 20,color = 'red',rotation = 0)
plt.title('圆',fontproperties = 'SimHei',size = 20)
```
