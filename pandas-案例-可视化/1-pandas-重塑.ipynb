```python
import numpy as np

import pandas as pd
from pandas import Series,DataFrame
```

```python
df = DataFrame(np.random.randint(0,150,size = (10,4)),columns=['Python','Math','En','Chinese'],
               index = pd.MultiIndex.from_product([['A','B','C','D','E'],['期中','期末']]))
df
```

```python
# 堆叠，列索引变成了行索引
stacked = df.stack()
stacked
```

```python
# unstack 行变成列
stacked.unstack(level=0)
```
