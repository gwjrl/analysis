```python
import numpy as np
import pandas as pd
from sqlalchemy import  create_engine
```

```python
engine = create_engine('mysql+pymysql://root:123456@127.0.0.1:3306/testdb?charset=utf8')
```

```python
detail = pd.read_sql_table('meal_order_detail1', con=engine)
detailPivot = pd.pivot_table(detail[['order_id', 'counts', 'amounts']], index='order_id')   # 当不指定聚合函数时会默认使用numpy.mean进行聚合运算
detailPivot.head()
```

```python
# 加入聚合函数之后
detailPivot = pd.pivot_table(detail[['order_id', 'counts', 'amounts']], index='order_id',aggfunc= np.sum) 
detailPivot.head()
```

```python
# 行索引分组
detailPivot2 = pd.pivot_table(detail[['order_id', 'dishes_name', 'counts', 'amounts']], index=['order_id', 'dishes_name'], aggfunc=np.sum)
detailPivot2.head()
```

```python
# 指定行列分组
detailPivot2 = pd.pivot_table(detail[['order_id', 'dishes_name', 'counts', 'amounts']], index='order_id', columns='dishes_name',aggfunc=np.sum)
detailPivot2.iloc[:5, :4] 
```

#### 对透视表中的缺失值进行填充

```python
detailPivot3 = pd.pivot_table(detail[['order_id', 'dishes_name', 'counts', 'amounts']], index='order_id', columns='dishes_name',aggfunc=np.sum, fill_value=0)
detailPivot3.iloc[:5, :4]
```

```python
# 汇总数据
detailPivot4 = pd.pivot_table(detail[['order_id', 'dishes_name', 'counts', 'amounts']], index='order_id', columns='dishes_name',aggfunc=np.sum,fill_value=0, margins=True)
detailPivot4.iloc[:5,:4]
```

#### 使用crosstab函数创建交叉表
###### 主要用于计算分组频率

```python
detailCross = pd.crosstab(index=detail['order_id'], columns=detail['dishes_name'], values=detail['counts'], aggfunc=np.sum)
detailCross.iloc[:5,:5]
```
