```python
# 剔除空值或相同的列
def dropful(data):
    beforlen = data.shape[1]
    coNull = data.describe().loc['count'] == 0
    for i in range(len(coNull)):
        if coNull[i]:
            data.drop(coNull.index[i], axis=1, inplace=True)
    afterlen = data.shape[1]
    return data
dropful(master)
    
```

```python
loginfo = pd.read_csv('../data/Training_LogInfo.csv', encoding='gbk')
loginfo
```

```python
# 分分组
loginfo['Listinginfo1'] = pd.to_datetime(loginfo['Listinginfo1'])
loginfo['date'] = [i.date() for i in loginfo['Listinginfo1']]
logroup = loginfo[['date', 'LogInfo2', 'LogInfo1']].groupby(by='date')
logroup.size().head()
```

```python
daymean = logroup.agg({'LogInfo2': np.min, 'LogInfo1': np.sum})
daymean
```
