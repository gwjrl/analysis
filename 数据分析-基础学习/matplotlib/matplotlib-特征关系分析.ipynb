```python
plt.rcParams['font.sans-serif'] = 'SimHei'   # 设置中文显示
plt.rcParams['axes.unicode_minus'] = False
```

```python
data = np.load('../data/国民经济核算季度数据.npz', allow_pickle=True)   # 获取数据
data
```

```python

name = data['columns']  # 提取其中的列索引columns数组视为数据标签
values = data['values']   # 提取其中的具体数值
plt.figure(figsize=(8,8))  # 设置画布
plt.scatter(values[:,0],values[:,2],marker="o",c='red')   # 绘制散点图
plt.xlabel('年份')
plt.ylabel('生产总值（亿元）')
plt.ylim((0,225000))   # 设置y轴范围
plt.xticks(range(0,70,4), values[range(0,70,4),1], rotation = 45)
plt.title('2000~2017年各季度国民生产总值散点图')   #添加标题
plt.legend(['第一产业','第二产业','第三产业'])     # 添加图例
plt.savefig('../tmp/2000~2017年各季度国民生产总值散点图.png')
plt.show()
```

#### 折线图

```python
# plt.plot(*args, scalex=True, scaley=True, data=None, **kwargs) 
# 折线函数plot() 中的常用参数同scatter是一致的
```

```python
plt.figure(figsize=(8,8))   # 设置画布
plt.plot(values[:,0], values[:,2], color = 'r', linestyle = '--')
plt.xlabel('年份')   # 添加横轴标签
plt.ylabel('生产总值(亿元)')   # 添加y轴名称
plt.ylim((0,225000))
plt.xticks(range(0,70,4),values[range(0,70,4),1], rotation=45)
plt.title('2000-2017年各季度国民生产总值折线图')   # 添加标题
plt.savefig("../tmp/2000-2017年各季度国民生产总值折线图.png")
plt.show()
```

```python
plt.figure(figsize=(8,8))   # 设置画布
plt.plot(values[:,0], values[:,2], color = 'r', linestyle = '--',marker="o")   # 绘制点线图
plt.xlabel('年份')   # 添加横轴标签
plt.ylabel('生产总值(亿元)')   # 添加y轴名称
plt.ylim((0,225000))
plt.xticks(range(0,70,4),values[range(0,70,4),1], rotation=45)
plt.title('2000-2017年各季度国民生产总值点线图')   # 添加标题
plt.savefig("../tmp/2000-2017年各季度国民生产总值点线图.png")
plt.show()
```

```python
plt.figure(figsize=(8,8))   # 设置画布
plt.plot(values[:,0], values[:,3], 'bs-',
        values[:,0], values[:,4], 'ro-.',
        values[:,0],values[:,5],'gH--')  # 折线散点图
plt.xlabel('年份')   # 添加横轴标签
plt.ylabel('生产总值(亿元)')   # 添加y轴名称
plt.ylim((0,225000))
plt.xticks(range(0,70,4),values[range(0,70,4),1], rotation=45)
plt.title('2000-2017年各季度国民生产总值折线图')   # 添加标题
plt.legend(['第一产业','第二产业','第三产业'])
plt.savefig("../tmp/2000-2017年各季度国民生产总值折线散点图.png")
plt.show()
```
