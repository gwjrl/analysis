```python
import numpy as np 
import matplotlib.pyplot as plt   #导入绘图模块
```

```python
data = np.arange(0,1.1,0.01)  # 创建数组
data
```

```python
plt.title('lines')   # 添加标题
plt.xlabel('x')  # 添加x轴名称
plt.ylabel('y')
plt.xlim((0,1))  # 确定x,y轴的范围
plt.xticks([0,0.2,0.4,0.6,0.8,1])  # 规定，xy轴的刻度
plt.yticks([0,0.2,0.4,0.6,0.8,1])
plt.plot(data,data**2)   # 添加x^2 曲线
plt.plot(data, data**4)
plt.legend(['y=x^2', 'y=x^4'])
plt.savefig('../tmp/y=x^2.png')
plt.show()
```

#### 包含子图绘制的基础语法

```python
rad = np.arange(0, np.pi*2,0.01)
# 第一幅子图
p1 = plt.figure(figsize=(8,6),dpi=80)  # 确定画布大小
ax1 = p1.add_subplot(2,1,1)   # 创建两行1列的子图
plt.title("lines")
plt.xlabel('x')
plt.ylabel('y')
plt.xlim((0,1))
plt.ylim((0,1))
plt.xticks([0,0.2,0.4,0.6,0.8,1])
plt.yticks([0,0.2,0.4,0.6,0.8,1])
plt.plot(data,data**2)   # 添加x^2 曲线
plt.plot(data, data**4)
plt.legend(['y=x^2', 'y=x^4'])

# 第二幅子图

ax2 = p1.add_subplot(2,1,2)
plt.title('sin/cos')
plt.xlabel('rad')
plt.ylabel('value')
plt.xlim((0, np.pi*2))
plt.ylim((-1,1))
plt.xticks([0, np.pi/2,np.pi,np.pi*1.5,np.pi*2])
plt.yticks([-1, -0.5, 0, 0.5, 1])
plt.plot(rad, np.sin(rad))
plt.plot(rad, np.cos(rad))
plt.legend(['sin','cos'])
plt.savefig('../tmp/sincos.png')
plt.show()
```

#### 设置动态的rc参数
#####  使用rc配置文件来自定义图形的各种默认属性，被称为rc参数，是可以动态改变的。同时所有存储在字典变量中的rc参数都被称为rcParams

```python
x = np.linspace(0, np.pi*4)    # 生成x轴数据
y = np.sin(x)   # 生成y轴数据
plt.plot(x,y,label = "$sin(x)$")   # 绘图
plt.title("sin")
plt.savefig('../tmp/默认sin曲线.png')
plt.show()
```

```python
# 修改rc参数后的图
plt.rcParams['lines.linestyle'] = '-.'  # 设置显示样式
plt.rcParams['lines.linewidth'] = 3   # 设置宽度
plt.plot(x,y,label='$sin(x)$')   # 绘制三角函数
plt.title('sin')
plt.savefig('../tmp/修改rc参数后sin曲线.png')
plt.show()
```
