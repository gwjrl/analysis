```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.rcParams['font.sans-serif'] = 'SimHei'    # 设置中文显示
plt.rcParams['axes.unicode_minus'] = False    # 
```

```python

data = np.load('../data/国民经济核算季度数据.npz', allow_pickle=True)
name = data['columns']   # 提取其中的columns数组
values = data['values']  # 提取其中的values数组，视为数据的存在位置
label1 = ['第一产业', '第二产业', '第三产业']   # 刻度标签
label2 = ['农业','工业','建筑','批发','交通','餐饮','金融','房地产','其他']
```

#### 绘制直方图及子图


```python
p = plt.figure(figsize=(12,12))
# 1
ax1 = p.add_subplot(2,2,1)   # 创建子图并指定其位置
plt.bar(range(3), values[0,3:6],width = 0.5)   # 绘制直方图
plt.xlabel('产业')
plt.ylabel('s生产总值（亿元）')
plt.xticks(range(3),label1)
plt.title('2000年第一季度国民生产总值产业构成分布直方图')
# 2
ax2 = p.add_subplot(2,2,2)
plt.bar(range(9), values[-1,6:],width = 0.5)
plt.xlabel('产业')
plt.ylabel('s生产总值（亿元）')
plt.xticks(range(9),label2)
plt.title('2000年第一季度国民生产总值产业构成分布直方图')
```

#### 绘制饼图

```python
explode1 = [0.01,0.01,0.01]
explode2 = [0.01,0.01,0.01,0.01,0.01,0.01,0.01,0.01,0.01]
p = plt.figure(figsize=(12,12))
ax1 = p.add_subplot(2,2,1)
plt.pie(values[0,3:6],explode=explode1,labels=label1,autopct='%1.1f%%')
plt.title('2000年第一季度国民生产总值产业构成饼图')
# 2
ax1 = p.add_subplot(2,2,2)
plt.pie(values[0,6:],explode=explode2,labels=label2,autopct='%1.1f%%')
plt.title('2000年第一季度国民生产总值产业构成饼图')
```

#### 箱线图

```python
gdp = (list(values[:,3]),list(values[:,4]),list(values[:,5]))
gdp2 = ([list(values[:,i]) for i in range(6,15)])   # 各个行业标签
p = plt.figure(figsize=(8,8))
# 1
ax1 = p.add_subplot(2,1,1)
# 绘制箱线图
plt.boxplot(gdp,notch=True,labels=label1,meanline=True)
plt.title('2000~2017年各产业国民生产总值箱线图')
plt.ylabel('生产总值（亿元）')
# 2
ax1 = p.add_subplot(2,1,2)
# 绘制箱线图
plt.boxplot(gdp2,notch=True,labels=label2,meanline=True)
plt.title('2000~2017年各产业国民生产总值箱线图')
plt.xlabel('hasattr业')
plt.ylabel('生产总值（亿元）')
# 
plt.savefig('../tmp/国民生产总值分散情况箱线图.png')
plt.show()
```
