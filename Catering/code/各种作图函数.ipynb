```python
import numpy as np
import pandas as pd
from pandas  import DataFrame,Series
import matplotlib.pyplot as plt    # 作图函数
```

```python
plt.rcParams['font.sans-serif'] = ['SimHei']  # 让中文标签正常显示
plt.rcParams['axes.unicode_minus'] = False   # 用来正常显示负号
```

```python
# 线形图或直方图
x = np.linspace(0, 2*np.pi, 50)   # 此函数用于输入X坐标
y = np.sin(x)                     # 计算对应的正弦值
plt.plot(x, y, 'bp--')            # 字符参数表示控制图像格式为蓝色虚线
plt.show()                        # 专门显示图像
```

```python
# 饼图
labels = 'F', 'H', 'D', 'L'   # 用于定义标签
sizes = [15,30,45,10]         # 定义饼图每块比例
colors = ['yellowgreen', 'gold', 'red', 'lightcoral']  # 饼图每块颜色序列
explode = (0, 0.1, 0, 0)      # 突出显示，
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%',
       shadow=True, startangle=90)    # 绘制饼图
plt.axis('equal')  # 控制输出格式
plt.show()
```

```python
# 二维条形直方图
x = np.random.randn(1000)       # 1000个服从正态分布的随机数
plt.hist(x, 10)                 # 将1000个数据分为10组
plt.show()
```

```python
# 箱型图
x = np.random.randn(1000)       # 1000个服从正态分布的随机数
D = pd.DataFrame([x, x+1]).T      # 构造两列DataFrame对象
D.plot(kind = 'box')            # 利用Series内置方法作图
plt.show()
```

```python
# 绘制x或y轴的对数图形
x = pd.Series(np.arange(20))   # 创建原始数据
x.plot(label = u'原数据图', legend = True)
plt.show()
x.plot(logy = True, label = u'对数数据图', legend = True)
plt.show()
```

```python
# 绘制误差棒图
error = np.random.randn(10)   # 创建误差列
y = pd.Series(np.sin(np.arange(10)))   # 创建均值数列
y.plot(yerr = error)          # 绘制y轴误差图
plt.show()
```
