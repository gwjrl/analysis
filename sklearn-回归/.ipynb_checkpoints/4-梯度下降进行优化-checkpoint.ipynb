```python
import numpy as np

import matplotlib.pyplot as plt
```

```python
# 开口向上，最小值
# 凹函数
f = lambda x : (x - 2.5)**2 + 8.5*x - 11

x = np.linspace(-8,5,100)

y = f(x)

plt.plot(x,y)
```
